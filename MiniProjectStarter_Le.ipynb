**Section 2:  Data Preprocessing**





```python
#Clean up profile_df

#remove rows with NaN in 'gender' and 'income'
profile_df_clean = profile_df.dropna(subset=['gender', 'income'])
print("Before cleanup: ",len(profile_df))
print("After cleanup: ",len(profile_df_clean))
```

```python
#Convert become_member_on to datetime
profile_df_clean = profile_df_clean.copy()
profile_df_clean['became_member_on'] = pd.to_datetime(profile_df_clean['became_member_on'], format='%Y%m%d')
```

```python
#transactions_df: reformat the key:value {'offer id': '9b98b8c7a33c4b65b9aebfe6a799e6d9'} into just the string value '9b98b8c7a33c4b65b9aebfe6a799e6d9'
transcript_df_clean = transcript_df.copy()
def parse_on_colon(s):
    s = s.strip("{} ")                # remove curly braces and spaces
    if ":" not in s:
        return s
    # split only on the first colon
    parts = s.split(":", 1)
    value = parts[1].strip().strip("'\" ")
    return value

transcript_df_clean["value"] = transcript_df_clean["value"].apply(parse_on_colon)
```

```python
#confirm 'value' format
transcript_df_clean.head()
```

```python
#Join transcript_df_clean with profile_df_clean
transcript_profile_df = pd.merge(transcript_df_clean, profile_df_clean, left_on='person', right_on='id')
transcript_profile_df.drop(columns=['id'], inplace=True)
transcript_profile_df.head()
```
